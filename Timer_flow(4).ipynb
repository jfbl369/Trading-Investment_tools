# Trading tool
## Clock with economic news alert

Everyone knows that following an economic news release, the market can experience huge gaps or completely irrational movements due to volatility.I wanted a tool to help me avoid these specific moments.

In an intraday trading strategy, we have a tremendous amount of data, information, news, or market movements to monitor at the same time. I wanted to have a simple and ergonomic tool, that's why I developed this clock/alarm that notifies me (based on a predefined interval) of upcoming and past news to avoid taking positions at dangerous times.

Every day, I retrieve data automatically from Investing's economic calendar by webscrapping and with a preprocessing step to clean and prepare the data to be injected into the Excel file containing my clock.

The clock consists of three fields:
1. The current date
2. The current time
3. The upcoming economic event (according to the predefined interval)

By default, the interval is set to 15 minutes, which means that my clock will be highlighted in red 15 minutes before the precise time of the event and 15 minutes after the event.
